```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pathlib as pl
```

```python
df_train = pd.read_csv('data/train.csv')

df_train.shape
```

```python
for i in range(3):
    array = df_train.iloc[i][1:]
    label = df_train.iloc[i][0]

    print(f'This is a picture of a {label}.')

    array = array.tolist()

    width = 28
    height = 28

    reshaped = np.array(array).reshape(height, width)
    reshaped = (reshaped * 255).astype(np.uint8)

    plt.imshow(reshaped.squeeze(), cmap='gray')  # 'gray' colormap for grayscale images
    plt.title('Image Title')
    plt.colorbar()  # Add a colorbar for reference
    plt.show()
```
